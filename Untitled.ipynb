```python
#clean_text= text.replace("n", " ")
clean_text= text.replace("\\", " ")       
#clean_text= ''.join([c for c in clean_text if c != "'"])
clean_text = clean_text.split('\n')
clean_text = [value for value in clean_text if value != '']
clean_text = [value for value in clean_text if value != ' ']
clean_text
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
final_words = []
for each in clean_text:
    t = []
    s = ""
    t = each.split()
    for e in t:
        if e not in stopwords:
            s = s+e+" "
    final_words.append(s)
final_words
```

```python
textblob_sentiment=[]
for s in clean_text:
    txt= TextBlob(s)
    a= txt.sentiment.polarity
    b= txt.sentiment.subjectivity
    textblob_sentiment.append([s,a,b])
textblob_sentiment
```

```python
df_textblob = pd.DataFrame(textblob_sentiment, columns =['Sentence', 'Polarity', 'Subjectivity'])
```

```python
df_textblob.head()
```

```python
df_textblob.loc[:,'Sentence']
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
sentiment = SentimentIntensityAnalyzer()
t = []
for e in final_words:
    sent_1 = sentiment.polarity_scores(e)
    print(sent_1)
    t.append(sent_1)
df = pd.DataFrame(t)
df
```

```python
result = pd.concat([df_textblob, df], axis=1, join='inner')
result.rename(columns={"neg": "Negative"})
result.rename(columns={"neu": "Neutral"})
result.rename(columns={"pos": "Positive"})
```
